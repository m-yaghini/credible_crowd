```python
anywhere_blocklist = ['(Andrew Cline / Shutterstock, Inc.)',
                   '(Alan Freed / Shutterstock, Inc)',
                   '... from Russia Today, Moscow',
                   'ISIS is going to have fun flying this baby',
                   '...by Gordon Duff, VT Senior Editor',
                   '(Joseph Sohm / Shutterstock, Inc.)',
                   'President Rump enters Marine One on White House lawn',
                   'By Sajjad Shaukat for Veterans Today', 
                   "READ MORE: Trident whistleblower calls out MoD's 'lame attempt' to excuse nuke malfunctions",
                   "READ MORE: Trident nukes useless against today's actual security problems - CND report",
                   'Read more',
                   "... from  Russia Today, Moscow",
                   " VIDEO   Please Donate to The Common Sense Show",
                   "by Jim W. Dean, VT Editor,   ...with Press TV,  Tehran"
                   ]

end_blocklist= ['Delivered by The Daily Sheeple', 'Chase Logan', 
                'Contributed by Organic and Healthy of Organic and Healthy.',
                'Please Donate to The Common Sense Show',
                "Gordon's Archives - 2008-2013",
                '__________',
                '(AP Photo)',
                'Shane Paulson',
                
               ]

expr_blocklist = ['MICHAEL SNYDER-', 'SHAUN BRADLEY-', 'RT/Moscow:', 'by Healthy Cures',
                 "RELATED POSTS Let's block ads! (Why?) Every Day is Earth Day Source:  http://feedproxy.google.com/~r/riseearth/KZKa/~3/hx8PpY5UXw4/clean-your-arteries-and-prevent-heart.htmlWe encourage you to Share our Reports, Analyses, Breaking News and Videos. Simply Click your Favorite Social Media Button and Share."]
```

```python
def clean_text(text):
    import unidecode
    # Convert to ASCII
    text_ = unidecode.unidecode(text)
    
    # Filter Blocked Expressions
    for expr in expr_blocklist:
        text_ = text_.replace(expr, '')
        
    # Clean blocked paragraphs and fix multiple line breaks
    cleaned = []
    for par in text_.split('\n'):
        par_cleaned = par.lstrip().strip()
        if par_cleaned in anywhere_blocklist:
            continue;
        if par_cleaned in end_blocklist:
            break;
        if len(par_cleaned)> 0:
            cleaned.append(par_cleaned)
    return '\n'.join(cleaned)
```

```python
# Clean text
final_DF.text = final_DF.text.apply(clean_text)
```

## Reading Time

```python
with_reading_time = final_DF.copy()
```

```python
def reading_time(row):
    words = row.text.split(' ') + row.title.split(' ')
#     print(words)
    read_time = (len(words)/200.0)*60  #time in seconds, assumes 200 words per minute reading speed.
    return read_time
```

```python
with_reading_time['reading_time'] = with_reading_time.apply(reading_time, axis=1)
```

```python
with_reading_time = with_reading_time[lambda x: x.reading_time < 200]
```

```python
final_DF_edited = final_DF_edited.reset_index()\
        .join(pd.DataFrame(honeypot_sentences, columns=['honeypot_sentences']))\
        .join(pd.DataFrame(honeypot_answers, columns=['honeypot_answers']))\
        .set_index('orig_index')
```

```python
final_DF_edited['honeypot_answers_hashed'] = final_DF_edited.honeypot_answers.apply(lambda x: hashlib.md5(x.encode()).hexdigest())
```

```python
def add_honeypot_row(row):
    import nltk
    _text = row['text']
    try:
        sentences = nltk.sent_tokenize(_text)
    except TypeError:
        print(row.title)
        print("\n")
        print(_text)
    chosen_sentence_idx = np.random.randint(low=int(len(sentences)*1/2), high=len(sentences)-1)
    chosen_sentence = sentences[chosen_sentence_idx]
    print(row['honeypot_sentences'].capitalize())
    _text = _text.replace(chosen_sentence, chosen_sentence + ' ' + row['honeypot_sentences'].capitalize())
    return _text
```

```python
final_DF_edited.text = final_DF_edited.apply(add_honeypot_row, axis=1)
```
